```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

analysis_data = {
    'TeamName': ['Bucks', 'Clippers', 'Lakers', 'Nets'],
    'Conference': ['East', 'West', 'West', 'East'],
    'Total_Salary': [73, 78, 76, 79],
    'Avg_Points': [23.5, 24.5, 23.5, 28.0]
}
df_analysis = pd.DataFrame(analysis_data)

df_analysis['Cost_per_Point'] = df_analysis['Total_Salary'] / df_analysis['Avg_Points']

sns.barplot(x='TeamName', y='Total_Salary', data=df_analysis)
plt.show()

sns.scatterplot(x='Total_Salary', y='Avg_Points', hue='Conference', data=df_analysis)
plt.title('Team Salary vs. Performance')
plt.xlabel('Total Salary (in Millions)')
plt.ylabel('Average Points')
plt.show()
```
