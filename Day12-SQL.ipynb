```python
# SQL 날짜 함수 - 데이터베이스에 저장된 날짜/시간 데이터에서 원하는 정보를 추출하거나 특정 형식으로 바꿀 때 사용함.
# 1. 날짜 정보 추출하기: DATE()
SELECT
    OrderID,
    DATE(OrderDaterTIme) AS OrderDate
FROM
    Orders;

# 2. 날짜를 원하는 형식으로 바꾸기: STRFTIME()은 날짜/시간 데이터를 내가 원하는 형식의 문자열로 만들어줌.
# %Y: 4자리 년도 / %m: 2자리 월 / %d: 2자리 일 / %w: 요일
SELECT
    STRFTIME('%Y-%m', OrderDateTime) AS OrderMonth,
    COUNT(OrderID) AS Num_Orders
FROM
    Orders
GROUP BY
    OrderMonth;

# 연습문제
UserActivity 테이블:
| LogID | UserID | Action      | Timestamp           |
| :---- | :----- | :---------- | :------------------ |
| 1     | 101    | 'login'     | 2025-09-20 08:00:00 |
| 2     | 102    | 'click'     | 2025-09-20 09:15:00 |
| 3     | 101    | 'logout'    | 2025-09-20 11:30:00 |
| 4     | 103    | 'login'     | 2025-09-21 10:00:00 |
| 5     | 102    | 'purchase'  | 2025-09-21 15:45:00 |
| 6     | 101    | 'login'     | 2025-10-01 09:00:00 |

SELECT
    COUNT(*) AS Total_Actions
FROM
    UserActivity
WHERE
    DATE(Timestamp) = '2025-09-20';

SELECT
    STARFTIME('%Y-%m, Timestamp) AS Month,
    COUNT(*) AS Total_Actions
FROM
    UserActivity
GROUP BY
    Month;
```
