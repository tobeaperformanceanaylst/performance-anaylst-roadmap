```python
#     - Subplot
# Subplot:  (Figure)    (Axes) ,     
# fig, axes = plt.subplots(nrows=1, ncols=2) * nrows=1, ncols=2 -  1 2  / axes[0]:    , axes[1]:    
# Subplot   plt.plot(), plt.title(),      - axes[0].plot(), axes[0].set_title()

import matplotlib.pyplot as plt

#  
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [78, 85, 92, 88, 110, 152, 145]

# 1. 1 2 (fig)  (axes) 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# 2.  (axes[0]) 
axes[0].plot(days, visitors, color='blue', marker='o')
axes[0].set_title('Weekly Visitor Trend (Line)') # set_title 
axes[0].set_xlabel('Day of the Week')           # set_xlabel 
axes[0].set_ylabel('Number of Visitors')      # set_ylabel 
axes[0].grid(True)

# 3.  (axes[1]) 
axes[1].bar(days, visitors, color='green')
axes[1].set_title('Weekly Visitor Comparison (Bar)')
axes[1].set_xlabel('Day of the Week')
axes[1].set_ylabel('Number of Visitors')

# 4.       
fig.suptitle('Weekly Visitor Analysis Dashboard', fontsize=16) #  
plt.tight_layout() #     
plt.show()
```

```python
import matplotlib.pyplot as plt

#   (x)
quarters = ['1Q', '2Q', '3Q', '4Q']

#   
math_scores = [80, 85, 92, 88]

#   
science_scores = [75, 82, 88, 95]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

axes[0].plot(quarters, math_scores)
axes[0].set_title('Math Score Trend')
axes[0].set_xlabel('quarters')
axes[0].set_ylabel('math_scores')
axes[0].grid(True)

axes[1].bar(quarters, science_scores)
axes[1].set_title('Science Score Comparison')
axes[1].set_xlabel('quarters')
axes[1].set_ylabel('science_scores')

plt.suptitle('Quarterly Academic Performance', fontsize=16)
plt.tight_layout()
plt.show()
```
