```python
# 여러 그래프 한번엥 그리기 - Subplot
# Subplot: 하나의 도화지(Figure)를 여러 개의 작은 구역(Axes)으로 나누어, 각 구역에 다른 그래프를 그림
# fig, axes = plt.subplots(nrows=1, ncols=2) * nrows=1, ncols=2 - 그래프를 1행 2열로 만듬 / axes[0]: 왼쪽 첫번 째 그래프, axes[1]: 오른쪽 두 번째 그래프
# Subplot을 사용할 때는 plt.plot()이나, plt.title()대신, 내가 그릴 칸을 직접 지정해야됨 - axes[0].plot(), axes[0].set_title()

import matplotlib.pyplot as plt

# 데이터 준비
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [78, 85, 92, 88, 110, 152, 145]

# 1. 1행 2열의 도화지(fig)와 그래프 칸(axes) 준비
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# 2. 왼쪽 그래프(axes[0]) 그리기
axes[0].plot(days, visitors, color='blue', marker='o')
axes[0].set_title('Weekly Visitor Trend (Line)') # set_title 사용
axes[0].set_xlabel('Day of the Week')           # set_xlabel 사용
axes[0].set_ylabel('Number of Visitors')      # set_ylabel 사용
axes[0].grid(True)

# 3. 오른쪽 그래프(axes[1]) 그리기
axes[1].bar(days, visitors, color='green')
axes[1].set_title('Weekly Visitor Comparison (Bar)')
axes[1].set_xlabel('Day of the Week')
axes[1].set_ylabel('Number of Visitors')

# 4. 전체 그림에 대한 정보 추가 및 보여주기
fig.suptitle('Weekly Visitor Analysis Dashboard', fontsize=16) # 전체 제목
plt.tight_layout() # 그래프들이 겹치지 않게 자동 조절
plt.show()
```

```python
import matplotlib.pyplot as plt

# 분기 데이터 (x축)
quarters = ['1Q', '2Q', '3Q', '4Q']

# 수학 분기별 성적
math_scores = [80, 85, 92, 88]

# 과학 분기별 성적
science_scores = [75, 82, 88, 95]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

axes[0].plot(quarters, math_scores)
axes[0].set_title('Math Score Trend')
axes[0].set_xlabel('quarters')
axes[0].set_ylabel('math_scores')
axes[0].grid(True)

axes[1].bar(quarters, science_scores)
axes[1].set_title('Science Score Comparison')
axes[1].set_xlabel('quarters')
axes[1].set_ylabel('science_scores')

plt.suptitle('Quarterly Academic Performance', fontsize=16)
plt.tight_layout()
plt.show()
```
