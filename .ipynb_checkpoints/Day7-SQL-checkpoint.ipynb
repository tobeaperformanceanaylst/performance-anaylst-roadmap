```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot

data = {
    'ice_cream_sales': [20, 25, 30, 38, 42],
    'sunglasses_sales': [30, 35, 40, 50, 55],
    'hot_chocolate_sales': [15, 12, 10, 5, 3]
}
df_sales = pd.DataFrame(data)

corr = df_sales['sunglasses_sales'].corr(df_sales['hot_chocolate_sales'])

sns.scatterplot(x='sunglasses_sales', y='hot_chocolate_sales', data=df_sales)
plt.show()
```
