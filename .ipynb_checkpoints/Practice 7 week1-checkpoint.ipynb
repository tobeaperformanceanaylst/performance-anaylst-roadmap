```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Title': ['The Action Hero', 'Galaxy Quest 3', 'Love in Paris', 'Midnight Shadow', 'Dino Planet', 'Silent Witness', 'Cosmic Warriors', 'Paris Again'],
    'Genre': ['Action', 'Sci-Fi', 'Romance', 'Thriller', 'Animation', 'Thriller', 'Sci-Fi', 'Romance'],
    'Release_Date': pd.to_datetime(['2024-01-15', '2024-02-20', '2024-03-08', '2024-04-17', '2024-05-25', '2024-06-30', '2024-07-19', '2024-08-22']),
    'Budget': [150, 200, 50, 80, 180, np.nan, 220, 45], #   ( )
    'Revenue': [380, 550, 120, 150, 450, 95, 610, 98], #   ( )
    'Audience_Score': [8.5, 9.1, 7.8, 8.2, 8.8, 6.9, 9.3, 7.5] #   (10 )
}
df_movies = pd.DataFrame(data)

#    
df_movies['Budget'] = df_movies['Budget'].fillna(df_movies.groupby('Genre')['Budget'].transform('mean'))  #   

#   
df_movies['Profit'] = df_movies['Revenue'] - df_movies['Budget']  # 'Profit' 

df_movies['ROI'] = (df_movies['Profit'] / df_movies['Budget']) * 100  # 'ROI' 

#    
avg_roi = df_movies.groupby('Genre')['ROI'].mean()  #  ROI 

top_roi = df_movies[(df_movies['Audience_Score'] >= 8.0) & (df_movies['Profit'] > 0)]  # 

# print(df_movies[['Title', 'ROI']]) #   

# 
sns.scatterplot(x='Budget', y='Revenue', hue='Genre', data=df_movies) #     
plt.show()
#       ,   ,    sci-fi     

sns.barplot(x='Genre', y='Profit', data=df_movies)  #      
plt.xticks(rotation=45)  # x    
plt.show()
#        Sci-Fi,    Thriller        .

sns.boxplot(x='Genre', y='Audience_Score', data=df_movies)
plt.show()
#           Thriller 
```
