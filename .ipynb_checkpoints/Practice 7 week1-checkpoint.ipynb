```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Title': ['The Action Hero', 'Galaxy Quest 3', 'Love in Paris', 'Midnight Shadow', 'Dino Planet', 'Silent Witness', 'Cosmic Warriors', 'Paris Again'],
    'Genre': ['Action', 'Sci-Fi', 'Romance', 'Thriller', 'Animation', 'Thriller', 'Sci-Fi', 'Romance'],
    'Release_Date': pd.to_datetime(['2024-01-15', '2024-02-20', '2024-03-08', '2024-04-17', '2024-05-25', '2024-06-30', '2024-07-19', '2024-08-22']),
    'Budget': [150, 200, 50, 80, 180, np.nan, 220, 45], # 제작 예산 (백만 달러)
    'Revenue': [380, 550, 120, 150, 450, 95, 610, 98], # 흥행 수익 (백만 달러)
    'Audience_Score': [8.5, 9.1, 7.8, 8.2, 8.8, 6.9, 9.3, 7.5] # 관객 평점 (10점 만점)
}
df_movies = pd.DataFrame(data)

# 데이터 정제 및 전처리
df_movies['Budget'] = df_movies['Budget'].fillna(df_movies.groupby('Genre')['Budget'].transform('mean'))  # 결측치 평균값으로 채우기

# 새로운 정보 생성
df_movies['Profit'] = df_movies['Revenue'] - df_movies['Budget']  # 'Profit'열 추가

df_movies['ROI'] = (df_movies['Profit'] / df_movies['Budget']) * 100  # 'ROI'열 추가

# 심층 분석 및 필터링
avg_roi = df_movies.groupby('Genre')['ROI'].mean()  # 장르별 ROI평균 

top_roi = df_movies[(df_movies['Audience_Score'] >= 8.0) & (df_movies['Profit'] > 0)]  # 필터링

# print(df_movies[['Title', 'ROI']]) # 원하는 열만 출력

# 시각화
sns.scatterplot(x='Budget', y='Revenue', hue='Genre', data=df_movies) # 예산과 수익의 관계를 보기위한 산점도
plt.show()
# 그래프의 점들이 점진적으로 위로 향하고 있는것으로 보아, 양의 상관계수를 띄며, 예산이 많이 사용되는 sci-fi같은 장르는 예산과 수익의 관계성을 보여준다

sns.barplot(x='Genre', y='Profit', data=df_movies)  # 장르별 평균 순수익 비교를 위한 막대그래프
plt.xticks(rotation=45)  # x축 라벨을 가독성을 위해 회전
plt.show()
# 장르별 평균 순수익이 가장 높게 책정되는 장르는 Sci-Fi이며, 가장 낮은 장르는 Thriller이고 신뢰구간 역시 가장 큰폭으로 데이터의 신뢰성이 높지는 않다.

sns.boxplot(x='Genre', y='Audience_Score', data=df_movies)
plt.show()
# 막대그래프와 거의 같은 분석결과를 보여주며 평간의 편차가 가장 큰 장르는 Thriller 이다
```
