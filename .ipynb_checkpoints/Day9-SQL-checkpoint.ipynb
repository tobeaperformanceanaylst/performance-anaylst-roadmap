```python
# JOIN 심화 - LEFT & RIGHT JOIN
# LEFT JOIN: 왼쪽 테이블은 모두 포함 (가장 일반적으로 많이 사용됨)
# - FROM절에 오는 왼쪽 테이블의 모든 데이터를 일단 보여준 후 ON 조건에 맞는 데이터가 오른쪽 테이블에 있다면 붙이고, 없다면 그 자리를 NULL로 채움
SELECT # 내가 보고 싶은 각 테이블의 열 지정
    Students.Name,
    Clubs.ClubName
FROM # 원하는 테이블 불러오기
    Students
LEFT JOIN
    Clubs
ON
    Students.StudentID = Clubs.StudentID

# RIGHT JOIN: 오른쪽 테이블은 모두 포함
# - JOIN절에 오는 오른쪽 테이블의 모든 데이터를 일단 다 보여줌.
SELECT
    Students.Name,
    Clubs.ClubName
FROM
    Students
RIGHT JOIN
    Clubs
ON
    Students.StudentID = Clubs.StudentID;
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Pandas와 Matplotlib으로 시각화하기
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Campaign': ['A', 'B', 'A', 'B', 'A', np.nan, 'B'],
    'Clicks': [10, 15, 8, 18, 12, np.nan, 20]
}
df_campaign = pd.DataFrame(data)

# Clicks가 없는(캠페인 미참여) 고객은 0으로 처리
df_campaign['Clicks'].fillna(0)


# 1. 데이터 요약
# 캠페인별 평균 클릭 수 계산
avg_clicks_per_campaign = df_campaign.groupby('Campaign')['Clicks'].mean()

# 캠페인별 참여 고객 수 계산 (NaN이 아닌 값만 카운트)
participants_per_campaign = df_campaign['Campaign'].value_counts()


# 2. 시각화 (Subplot)
# 1행 2열의 도화지(fig)와 그래프 칸(axes) 준비
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# 왼쪽 그래프 (axes[0]): 캠페인별 평균 클릭 수
sns.barplot(x=avg_clicks_per_campaign.index, y=avg_clicks_per_campaign.values, ax=axes[0])
axes[0].set_title('Average Clicks per Campaign')
axes[0].set_xlabel('Campaign')
axes[0].set_ylabel('Average Clicks')

# 오른쪽 그래프 (axes[1]): 캠페인별 참여 고객 수
sns.barplot(x=participants_per_campaign.index, y=participants_per_campaign.values, ax=axes[1])
axes[1].set_title('Number of Participants per Campaign')
axes[1].set_xlabel('Campaign')
axes[1].set_ylabel('Number of Participants')

# 전체 대시보드 제목 및 레이아웃 설정
fig.suptitle('Campaign Performance Analysis', fontsize=16)
plt.tight_layout()
plt.show()
```
