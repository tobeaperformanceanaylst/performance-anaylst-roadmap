```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'CustomerID': range(1, 16),
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male'],
    'Age': [25, 31, 19, 35, 42, 28, 33, 45, 22, 38, 29, 34, 41, 18, 36],
    'Favorite_Category': [' Electronics ', ' Fashion', 'Books', 'Home Goods ', ' Electronics', 'Fashion', 'Books', ' Home Goods', 'Electronics', ' Fashion', 'Books', 'Home Goods', 'Fashion', ' Books ', 'Electronics'],
    'Spent_Last_Month': [350, 420, 150, 280, 510, 380, 210, 310, 450, 390, 250, 290, 410, 180, 480], # 지난달 소비 금액 ($)
    'Membership_Level': ['Gold', 'Silver', 'Bronze', 'Silver', 'Gold', 'Silver', 'Bronze', 'Gold', 'Gold', 'Silver', 'Bronze', 'Silver', 'Gold', 'Bronze', 'Gold']
}
df_shop = pd.DataFrame(data)

# 데이터 정제
df_shop['Favorite_Category'] = df_shop['Favorite_Category'].str.strip() # 문자열 공백 처리

# 새로운 정보 생성
age_bins = [10, 19, 29, 39, 49] # 나이대 별로 구별
age_labels = ['10-19', '20-29', '30-39', '40-49'] # 그룹별 이름
df_shop['Age_Group'] = pd.cut(df_shop['Age'], bins=age_bins, labels=age_labels) # age group 만들기

# 심층분석
pivot_table = pd.pivot_table(df_shop, index='Age_Group', columns='Gender', values='Spent_Last_Month', aggfunc='mean', observed=False)
# print(pivot_table)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title('Average Monthly Spend by Age Group and Gender')
plt.ylabel('Age Group')
plt.xlabel('Gender')
plt.show()
```

```python
# 데이터 결합 - pd.merge: 두 개의 데이터프레임에서 공통된 값을 가진 열(key)을 기준으로 데이터를 연결
# 1. Inner Join(내부 결합): 두 데이터프레임에 모두 존재하는 Key만 남김. * how='inner'(기본값)
# ex) pd.merge(df_users, df_purchases, how='inner', on='UserID')

# 2. Left Join(왼쪽 결합): 왼쪽 데이터프레임을 기준으로, 모든 행을 유지하고 오른쪽에서 일치하는 데이터를 가져와 붙임. 일치하는 데이터가 없으면 그 자리는 NaN으로 표시됨. * how='left'
# ex) pd.merge(df_users, df_purchases, how='left', on=UserID')

# 3. Right Join(오른쪽 결합): 오른쪽 데이터프레임을 기준으로 모든 행을 유지함. * how='right'
# ex) pd.merge(df_users, df_purchases, how='right', on='UserID')

# 4. Outer Join(외부 결합): 어느 한쪽에라도 존재하는 모든 Key를 남김. * how='outer'
# ex) pd.merge(df_uers, df_purchases, how='outer', on=UserID)

# 데이터 결합 - pd.concat: 여러 개의 데이터프레임을 위아래(수직) 또는 양옆(수평)으로 붙이는 기능. * merge처럼 공통된 Key를 찾지 않고, 지정한 축을 기준으로함.
# 컬럼 구조가 똑같은 여러 개의 파일을 하나로 합칠 때 사용됨
# 1. 위아래로 이어 붙이기(수직 결합) - axis=0(기본값)
# ex) pd.concat([df_ql, df_q2], ignore_index=True) * ignore_index=True - 인덱스를 0부터

# 2. 양옆으로 이어 붙이기(수평 결합) - axis=1
# ex) pd.concat([df_left, df_right], axis=1)

import pandas as pd

# 데이터 1: 직원의 개인 정보
df_personal_info = pd.DataFrame({
    'EmployeeID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Hire_Date': ['2021-05-10', '2022-03-15', '2021-11-20', '2023-01-05', '2022-08-01']
})

# 데이터 2: 직원의 업무 정보
df_work_info = pd.DataFrame({
    'EmployeeID': [101, 102, 104, 106],
    'Department': ['HR', 'Marketing', 'Engineering', 'Sales'],
    'Salary': [7000, 8500, 9500, 8800]
})

active_employee = pd.merge(df_personal_info, df_work_info, how='inner', on='EmployeeID')
standard_personal = pd.merge(df_personal_info, df_work_info, how='left', on='EmployeeID')

# 서울 본사 직원 명단
df_seoul = pd.DataFrame({
    'EmployeeID': [201, 202, 203],
    'Name': ['Frank', 'Grace', 'Henry'],
    'Team': ['Marketing_A', 'HR_B', 'Sales_A']
})

# 부산 지사 직원 명단
df_busan = pd.DataFrame({
    'EmployeeID': [301, 302],
    'Name': ['Ivy', 'Jack'],
    'Team': ['Sales_B', 'Marketing_C']
})

employee_list = pd.concat([df_seoul, df_busan], ignore_index=True)
```

```python
# 피벗 테이블(Pivot Table): 테이터를 원하는 행과 열을 기준으로 데이터를 재구성하고 요약함
# 구성 요소: index, colums, values, aggfunc(values를 어떤 방법으로 계산할지 정함)
# ex) df_cafe.pivot_table(index='Day_of_week', columns='Weather', values='Total_Sales', aggfunc='mean')

import pandas as pd

data = {
    'Game': ['Cosmic Odyssey', 'Soccer Pro 2025', 'Dragon Blade', 'Speed Demons', 'Kingdom of Valor', 'Gridiron Kings', 'Starfall Chronicles', 'Final Goal'],
    'Platform': ['PS5', 'Xbox', 'Switch', 'PS5', 'Switch', 'Xbox', 'PS5', 'Switch'],
    'Genre': ['RPG', 'Sports', 'RPG', 'Racing', 'RPG', 'Sports', 'RPG', 'Sports'],
    'Sales': [850, 920, 780, 650, 1150, 880, 950, 750] # 판매량 (만 달러)
}
df_games = pd.DataFrame(data)

pivot_table = df_games.pivot_table(index='Genre',
                                   columns='Platform',
                                   values='Sales',
                                   aggfunc='sum'
                                  )
pivot_table2 = df_games.pivot_table(index='Genre',
                                   columns='Platform',
                                   values='Sales',
                                   aggfunc='mean'
                                  )
```
