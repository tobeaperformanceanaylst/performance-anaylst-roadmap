```python
# Matplotlib: Python        
# import matplotlib.pyplot as plt
# plt.figure() -  :    .   
# plt.plot() / plt.bar()  -  :    (,  ) . x y  
# plt.title(), plt.xlabel(), plt.ylabel() -  :  , x y  
# plt.show() - :    

# 1.  (Line Plot) -       
import matplotlib.pyplot as plt
#   ( 10,  5 )
plt.figure(figsize=(10, 5))

#   
months = [1, 2, 3, 4, 5, 6] # x 
sales = [100, 120, 150, 130, 160, 155] # y 

#   
plt.plot(months, sales)

#  
plt.title('Monthly Sales Trend') # 
plt.xlabel('Month') # x 
plt.ylabel('Sales') # y 

# 
plt.show()

# 2. (Bar Plot) -      

# 1.  
plt.figure(figsize=(10, 5))

#   
fruits = ['Apple', 'Banana', 'Cherry'] # x 
counts = [30, 55, 40] # y 

# 2.  
plt.bar(fruits, counts)

# 3.  
plt.title('Fruit Counts')
plt.xlabel('Fruit')
plt.ylabel('Count')

# 4. 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#   (x)
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
#    (y)
visitors = [78, 85, 92, 88, 110, 152, 145]

plt.figure()
plt.plot(days, visitors)
plt.title('Weekly Visitor Trend')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Visitors')
plt.show()

plt.figure()
plt.bar(days, visitors)
plt.title('Daily Visitor Comparison')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Visitors')
plt.show()
```

```python
#  (Styling) - ,  ,         
# color:  , linestyle:  , marker:       , label:  
# * label     plt.legend()   .
import matplotlib.pyplot as plt

#  
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
actual_visitors = [78, 85, 92, 88, 110, 152, 145] #  
expected_visitors = [70, 80, 90, 100, 110, 140, 150] #  

# 1.  
plt.figure(figsize=(10, 6))

# 2.    ( )
#   :  ,  
plt.plot(days, actual_visitors, color='blue', marker='^', linestyle='-', label='Actual')

#   :  , x 
plt.plot(days, expected_visitors, color='red', marker='x', linestyle='--', label='Expected')

# 3.  
plt.title('Actual vs. Expected Visitors')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Visitors')
plt.grid(True) #    

# 4. (Legend) 
plt.legend()

# 5. 
plt.show()
```

```python
import matplotlib.pyplot as plt

#   (x)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

# ''    (kg)
sales_ethiopia = [22, 25, 30, 35, 42, 38]

# ''    (kg)
sales_colombia = [30, 32, 28, 38, 45, 50]

plt.figure(figsize=(10, 6))

plt.plot(months, sales_ethiopia, color='green', linestyle='--', marker='o', label='Ethiopia')

plt.plot(months, sales_colombia, color='brown', linestyle='-', marker='*', label='Colombia')

plt.title('Monthly Coffee Bean Sales')
plt.xlabel('Month')
plt.ylabel('Sales(kg)')
plt.grid(True)

plt.legend()

plt.show()
```
