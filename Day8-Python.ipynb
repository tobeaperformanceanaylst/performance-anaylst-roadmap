```python
# Matplotlib: Python에서 데이터를 그래프나 차트로 그려주는 가장 기본적이면서도 강력한 라이브러리
# import matplotlib.pyplot as plt
# plt.figure() - 도화지 준비하기: 그림을 그릴 도화지를 준비함. 크기 조절 가능
# plt.plot() / plt.bar() 등 - 그림 그리기: 도화지 위에 원하는 그래프(선, 막대 등)을 그림. x축과 y축 데이터 지정
# plt.title(), plt.xlabel(), plt.ylabel() - 정보 추가하기: 그래프의 제목, x축과 y축의 이름 붙이기
# plt.show() - 보여주기: 완성된 그림을 화면에 보여줌

# 1. 선 그래프(Line Plot) - 시간에 따른 데이터의 변화난 추세를 보여줄 떼
import matplotlib.pyplot as plt
# 도화지 준비 (가로 10, 세로 5 크기)
plt.figure(figsize=(10, 5))

# 그래프에 사용할 데이터
months = [1, 2, 3, 4, 5, 6] # x축 데이터
sales = [100, 120, 150, 130, 160, 155] # y축 데이터

# 선 그래프 그리기
plt.plot(months, sales)

# 정보 추가하기
plt.title('Monthly Sales Trend') # 제목
plt.xlabel('Month') # x축 이름
plt.ylabel('Sales') # y축 이름

# 보여주기
plt.show()

# 2. 막대그래프(Bar Plot) - 카테고리별 데이터의 크기를  비교할 때

# 1. 도화지 준비
plt.figure(figsize=(10, 5))

# 그래프에 사용할 데이터
fruits = ['Apple', 'Banana', 'Cherry'] # x축 데이터
counts = [30, 55, 40] # y축 데이터

# 2. 막대그래프 그리기
plt.bar(fruits, counts)

# 3. 정보 추가하기
plt.title('Fruit Counts')
plt.xlabel('Fruit')
plt.ylabel('Count')

# 4. 보여주기
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 요일 데이터 (x축)
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
# 일일 방문객 수 (y축)
visitors = [78, 85, 92, 88, 110, 152, 145]

plt.figure()
plt.plot(days, visitors)
plt.title('Weekly Visitor Trend')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Visitors')
plt.show()

plt.figure()
plt.bar(days, visitors)
plt.title('Daily Visitor Comparison')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Visitors')
plt.show()
```

```python
# 그래프 꾸미기(Styling) - 색상, 선 스타일, 범례 등을 추가하여 정보를 더 명확하게 전달할 수 있음
# color: 색상 지정, linestyle: 선 모양, marker: 선 그래프의 각 데이터 지점에 표식 추가, label: 데이터의 이름표
# * label을 화면 표시하기 위해서는 반드시 plt.legend() 함수를 함께 사용해야함.
import matplotlib.pyplot as plt

# 데이터 준비
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
actual_visitors = [78, 85, 92, 88, 110, 152, 145] # 실제 방문객
expected_visitors = [70, 80, 90, 100, 110, 140, 150] # 예상 방문객

# 1. 도화지 준비
plt.figure(figsize=(10, 6))

# 2. 여러 그래프 그리기 (스타일 적용)
# 실제 방문객 그래프: 파란색 실선, 동그라미 표식
plt.plot(days, actual_visitors, color='blue', marker='^', linestyle='-', label='Actual')

# 예상 방문객 그래프: 빨간색 점선, x자 표식
plt.plot(days, expected_visitors, color='red', marker='x', linestyle='--', label='Expected')

# 3. 정보 추가하기
plt.title('Actual vs. Expected Visitors')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Visitors')
plt.grid(True) # 배경에 격자 무늬 추가

# 4. 범례(Legend) 표시하기
plt.legend()

# 5. 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt

# 월 데이터 (x축)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

# '에티오피아' 원두 월별 판매량 (kg)
sales_ethiopia = [22, 25, 30, 35, 42, 38]

# '콜롬비아' 원두 월별 판매량 (kg)
sales_colombia = [30, 32, 28, 38, 45, 50]

plt.figure(figsize=(10, 6))

plt.plot(months, sales_ethiopia, color='green', linestyle='--', marker='o', label='Ethiopia')

plt.plot(months, sales_colombia, color='brown', linestyle='-', marker='*', label='Colombia')

plt.title('Monthly Coffee Bean Sales')
plt.xlabel('Month')
plt.ylabel('Sales(kg)')
plt.grid(True)

plt.legend()

plt.show()
```
