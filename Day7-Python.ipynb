```python
# Pandas 기본 통계
# 1. 분산(.var)과 표준편차(.std) - 두 지표 모두 데이터가 평균으로부터 얼마나 흩어져 있는지(산포도)를 나타냄
#  - 표준편차(.std)가 크면 클수록 데이터가 넓게 펴져있고, 작을 수록 평균에 밀집해 있음

# 2. 상관관계(.corr) - 두 변수가 얼마나 관련이 있는지를 -1부터 1 사이의 숫자로 나타냄
#  - 1에 가까울수록: 하나가 증가할 때 다른 하나도 증가하는 강한 양의 관계
#  - -1에 가까울수록: 하나가 증가할 때 다른 하나는 감소하는 강한 음의 관계
#  - 0에 가까울수록: 두 변수 사이에 관계가 거의 없음

import pandas as pd

data = {
    'study_hours': [2, 3, 5, 7, 8],
    'test_score': [65, 70, 80, 88, 92]
}
df = pd.DataFrame(data)

# 각 열의 표준편차 계산
print("--- 표준편차 ---")
print(df.std())

# 두 열 사이의 상관관계 계산
print("\n--- 상관관계 ---")
print(df.corr())
```

```python
import pandas as pd

data = {
    'ice_cream_sales': [20, 25, 30, 38, 42],
    'sunglasses_sales': [30, 35, 40, 50, 55],
    'hot_chocolate_sales': [15, 12, 10, 5, 3]
}
df_sales = pd.DataFrame(data)

print(df_sales.std())
print(df_sales.corr())
```

```python
# Seaborn으로 통계 시작화하기
# Seaborn: Matplotlib을 기반으로 더 아름답고 통계적인 차트를 쉽게 그릴 수 있게 해주는 라이브러리.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt # 차트를 화면에 보여주기 위함

# 1. barplot: 막대그래프 - 카테고리별 수치를 비교할 때 가장 기본적인 차트
data = {
    'department': ['Sales', 'HR', 'IT'],
    'salary': [75000, 52500, 87500]
}
df_dept = pd.DataFrame(data)

# barplot 그리기
sns.barplot(x='department', y='salary', data=df_dept)
plt.show() # 차트를 화면에 표시

# 2. boxplot: 박스플롯 - 데이터의 분포를 한눈에 보여주는 강력한 차트. .describe()로 본 통계치를 시각화한 것과 같음 (최솟값, 최댓값, 중앙값, 사분위스, 이상치)
# 1교시 때 사용했던 판매량 데이터
data = {
    'ice_cream_sales': [20, 25, 30, 38, 42],
    'sunglasses_sales': [30, 35, 40, 50, 55],
    'hot_chocolate_sales': [15, 12, 10, 5, 3]
}
df_sales = pd.DataFrame(data)

# boxplot 그리기
sns.boxplot(data=df_sales)
plt.show()

# 3. scatterplot: 산점도 - .corr()로 계산한 상관관계를 눈으로 직접 확인하는 차트. 두 숫자 데이터 사이의 관계를 점으로 나타냄
# scatterplot 그리기 (x축: 아이스크림 판매량, y축: 선글라스 판매량)
sns.scatterplot(x='ice_cream_sales', y='sunglasses_sales', data=df_sales)
plt.show()
# * 점이 우상향하면 양의 관계, 우하향하면 음의 관계를 의미
```
