```python
# 히스토그램(Histogram) - 히스토그램은 특정 데이터 값의 빈도수를 막대그래프 형태로 보여줌. 즉, 어떤 값의 범위에 데이터가 얼마나 많이 몰려있는지 쉽게 알 수 있음.
# Seaborn으로 그리기: sns.histplot()

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 가상의 고객 나이 데이터
ages = np.random.randint(20, 60, size=100)

# 히스토그램 그리기
sns.histplot(data=ages, bins=10) # bins는 막대의 개수를 의미
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.show()

# 박스플록 (Box Plot) - 데이터의 최솟값, 최댓값, 중앙값, 그리고 사분위수를 하나의 상자 그림으로 요약하여 보여줌. 데이터의 중심 경향, 퍼진 정도, 이상치 여부를 파악 가능함.
# Seaborn으로 그리기: sns.boxplot()

import pandas as pd

# 영화 평점 데이터
data = {
    'Genre': ['Action', 'Action', 'Action', 'Comedy', 'Comedy', 'Comedy', 'Drama', 'Drama', 'Drama'],
    'Score': [8.5, 7.9, 9.1, 6.5, 7.1, 8.2, 9.0, 9.2, 9.4]
}
df_movies = pd.DataFrame(data)

# 장르별 평점 분포를 박스플롯으로 그리기
sns.boxplot(x='Genre', y='Score', data=df_movies)
plt.title('Movie Score Distribution by Genre')
plt.show()
```

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 랜덤 시드 고정 (매번 같은 결과가 나오도록)
np.random.seed(42)

# 상품 데이터 생성
data = {
    'Category': ['Electronics'] * 50 + ['Fashion'] * 50 + ['Books'] * 50,
    'Price': np.concatenate([
        np.random.normal(loc=300, scale=80, size=50), # 전자기기 가격 (평균 300)
        np.random.normal(loc=80, scale=30, size=50),   # 패션 상품 가격 (평균 80)
        np.random.normal(loc=25, scale=10, size=50)    # 책 가격 (평균 25)
    ])
}
df_products = pd.DataFrame(data)
df_products['Price'] = df_products['Price'].round(2) # 소수점 2자리까지 반올림

# 전체 상품의 히스토그램
sns.histplot(x='Price', data=df_products)
plt.title('Overall Product Price Distribution')
plt.show()

# 카테고리별 가격의 분포 박스플롯
sns.boxplot(x='Category', y='Price', data=df_products)
plt.title('Price Distribution by Category')
plt.show()

# 평균적으로 비싼 제품은 전자제품이며, 가격 범위 역시 큰 폭을 보여준다.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'OS': ['iOS', 'iOS', 'Android', 'iOS', 'Android', 'Android', 'iOS', 'Android', 'iOS', 'Android', 'iOS', 'Android', 'iOS', 'Android'],
    'Screen_Time': [180, 210, 150, 200, 240, 130, 190, 280, 205, 110, 195, 310, 185, 160] # 일일 앱 사용 시간 (분)
}
df_usage = pd.DataFrame(data)

# 전체 사용자 사용 시간 분포 - 히스토그램
sns.histplot(x='Screen_Time', data=df_usage)
plt.title('Overall User Screen Time Distribution')
plt.show()

# OS별 사용 시간 비교 - 박스플롯
sns.boxplot(x='OS', y='Screen_Time', data=df_usage)
plt.title('Screen Time Distribution by OS')
plt.show()

# IOS 사용자의 스크린 타임이 더 일관적이다.
```
