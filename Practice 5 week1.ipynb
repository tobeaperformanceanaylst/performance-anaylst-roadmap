```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Player': ['LeBron James', 'Kevin Durant', 'Stephen Curry', 'Nikola Jokic', 'Joel Embiid', 'Luka Doncic', 'Jayson Tatum', 'Zion Williamson'],
    'Position': ['F', 'F', 'G', 'C', 'C', 'G', 'F', 'F'],
    'Points': [28.9, 29.1, 29.4, 24.5, 33.1, 32.4, 30.1, 26.0], # 경기당 득점
    'Rebounds': [8.3, 6.7, 6.1, 11.8, 10.2, 8.6, 8.8, 7.0], # 경기당 리바운드
    'Assists': [6.8, 5.0, 6.3, 9.8, 4.2, 8.0, 4.6, 4.6], # 경기당 어시스트
    'Experience': [20, 15, 14, 8, np.nan, 5, 6, 4], # NBA 경력 (년)
    'Salary': [44.4, 44.1, 48.0, 33.0, 33.6, np.nan, 30.3, 13.5] # 연봉 (백만 달러)
}
df_players = pd.DataFrame(data)

# 데이터 정제
avg_salary = df_players[df_players['Position'] == 'G']['Salary'].mean()
df_players['Salary'] = df_players['Salary'].fillna(avg_salary)

med_exp = df_players['Experience'].median()
df_players['Experience'] = df_players['Experience'].fillna(med_exp)

# 새로운 정보 생성
df_players['Overall_Score'] = df_players['Points'] + df_players['Rebounds'] + df_players['Assists']
df_players['Value_Score'] = df_players['Overall_Score'] / df_players['Salary']

# 심층 분석
top2_player = df_players[df_players['Position'] == 'F'].sort_values('Value_Score', ascending=False).head(2)

print(top2_player[['Player', 'Value_Score']])

# 통계 및 고급 시각화
corr_matrix = df_players['Experience'].corr(df_players['Salary'])

print(corr_matrix)

sns.scatterplot(x='Overall_Score', y='Salary', hue='Position', data=df_players)
plt.show()

sns.barplot(x='Position', y='Salary', data=df_players)
plt.show()
```
